```python
import csv
import json
```

```python
with open('restaurants.csv') as scraped_data:
    id = 0
    reader = csv.reader(scraped_data)
    for row in reader:
          index_info = {"index" : { "_index": "restaurants", "_id" : str(id)}}
          record = {"restaurant_id": str(row[0]), "cuisine": str(row[2])}
          id += 1
          with open('restaurants.json', 'a') as opensearch_file:
              json.dump(index_info, opensearch_file)
              opensearch_file.write('\n')
              json.dump(record, opensearch_file)
              opensearch_file.write('\n')

#delete the first row
```
